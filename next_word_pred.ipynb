```python
input_sequences = []

[[tokenized_sentence[:i+1] for i in range(1, len(tokenizer.texts_to_sequences([sentence])[0]))] tokenizer.texts_to_sequences([sentence])[0] for sentence in faqs.split('\n')]
```

```python
input_sequences = []
for sentence in faqs.split('\n'): # new line
    tokenized_sentence = tokenizer.texts_to_sequences([sentence])[0]
    print(tokenized_sentence)

    for i in range(1, len(tokenized_sentence)):
        input_sequences.append(tokenized_sentence[:i+1])
```

```python
input_sequences # about the 
                # about the programme
                # about the programme what 
                # about the programme what is the
                # about the programme what is the course 
                # about the programme what is the course fee for
                # about the programme what is the course fee for data 
```

```python
mx_len = max([len(x) for x in input_sequences])
mx_len
```

```python
from tensorflow.keras.preprocessing.sequence import pad_sequences
```

```python
pdded_input_sequences = pad_sequences(input_sequences, maxlen=mx_len, padding='pre')
```

```python
pdded_input_sequences
```

```python
X = pdded_input_sequences[:,:-1] # numpy slicing all row and columns except last columns
```

```python
y = pdded_input_sequences[:,-1]
```

```python
X.shape
```

```python
y.shape
```

```python
from tensorflow.keras.utils import to_categorical   # one hot encoding
```

```python
y = to_categorical(y,num_classes=283) # vocab size 282 >> 0 to 283
```

```python
y.shape
```

```python
y[0]
```

```python
y[1]
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Embedding, LSTM
```

```python
model = Sequential()
model.add(Embedding(283, 100, input_length=56)) # vocab size=283, output embeding shape =100 (for each word vector) each row vector contain 56 words (max no of words in a sentence)
model.add(LSTM(150))
model.add(Dense(283, activation='softmax'))
```

for multi class(n) classification target is transform by label encoder then one hot encoding makes dummy
columns same number(n) of class create vector or row of n elements contain one significant value for current
class rest are zero so output layer also be set for n number of classes output node with softmax activation
which calculate the probability(prediction) of the current class . then compare with the actual class vector
or row.

```python
# In the context of Recurrent Neural Networks (RNNs), LSTM(150) refers to a Long Short-Term Memory (LSTM) layer with 150 units.
# Breakdown:
# LSTM: This is the type of RNN layer, which is capable of learning long-term dependencies in data.
# (150): This specifies the number of units (also known as cells or neurons) in the LSTM layer. In this case, there are 150 units.
# # LSTM Layer:
# An LSTM layer is a type of RNN layer that is designed to handle the vanishing gradient problem, which occurs when training RNNs with backpropagation through time.
# 150 Units:
# The number of units in an LSTM layer determines the capacity of the layer to learn and represent complex patterns in data. In general:
# More units can learn more complex patterns, but may also increase the risk of overfitting.
# Fewer units may be less prone to overfitting, but may not be able to learn complex patterns.
# In this case, the LSTM layer has 150 units, which is a moderate number that can learn relatively complex patterns in data.
```

```python
model.compile(loss='categorical_crossentropy', optimizer='adam',metrics=['accuracy'])
```

```python
model.summary()
```

```python
model.fit(X,y,epochs=100)
```

```python
text = 'mail'

#tokenize
token_text = tokenizer.texts_to_sequences([text])[0]
print(token_text)

#padding

token_pad = pad_sequences([token_text], maxlen=56, padding='pre')

print(token_pad)

#predict

prediction = model.predict(token_pad)[0]
print(prediction)

#get the word with max probability
out_put_text = np.argmax(prediction)
print(out_put_text)
```

```python
import time
text = "what is the fee"

for i in range(10):
  # tokenize
  token_text = tokenizer.texts_to_sequences([text])[0]
  # padding
  padded_token_text = pad_sequences([token_text], maxlen=56, padding='pre')
  # predict
  pos = np.argmax(model.predict(padded_token_text))

  for word,index in tokenizer.word_index.items():
    if index == pos:
      text = text + " " + word
      print(text)
      time.sleep(2)
```
